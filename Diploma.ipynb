```python
# !pip install pandas
!pip install sodapy

from sodapy import Socrata
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

client = Socrata("data.ny.gov", None)
results = client.get("5xaw-6ayf", limit=3000)
results_df = pd.DataFrame.from_records(results)
```

```python
results_df
```

```python
results_df[['A', 'B', 'C', 'D', 'E']] = results_df.winning_numbers.str.split(" ", expand = True)
results_df.columns=results_df.columns.str.replace(' ','')
results_df['draw_date'] = pd.to_datetime(results_df['draw_date'])
```

```python

results_df.drop(['winning_numbers'], axis=1)
results_df = results_df.loc[:, ['draw_date', 'A', 'B', 'C', 'D', 'E', 'mega_ball', 'multiplier']]
```

```python
results_df
```

```python
rslt_df = results_df[(results_df['E'].isin([66, 46, 7])) & (results_df['C'].isin([66, 46, 7])) & (results_df['D'].isin([66, 46, 7])) & (results_df['B'].isin([66, 46, 7])) & (results_df['A'].isin([66, 46, 7]))]
rslt_df
```

```python
results_df['multiplier'] = results_df['multiplier'].fillna(1)
results_df
```

```python
results_df = results_df.astype({"A":"int", "B":"int", "C":"int", "D":"int", "E":"int", "mega_ball":"int", "multiplier":"int", })
results_df.dtypes
```

```python

%pip install mlxtend --upgrade
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import fpmax

frequent_itemsets = apriori(records, min_support=0.001, use_colnames=True)
frequent_itemsets
```

Association rule mining

```python
res = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
res.drop(['leverage', 'conviction'], axis=1)
```

```python
import seaborn as sns
# Convert antecedents and consequents into strings
# res['antecedents'] = res['antecedents'].apply(lambda a: ','.join(list(a)))
# res['consequents'] = res['consequents'].apply(lambda a: ','.join(list(a)))

# Transform antecedent, consequent, and support columns into matrix
support_table = res.pivot(index='consequents', columns='antecedents', values='confidence')

plt.figure(figsize=(10,6))
sns.heatmap(support_table, annot=True, cbar=False)
b, t = plt.ylim() 
b += 0.5 
t -= 0.5 
plt.ylim(b, t) 
plt.yticks(rotation=0)
plt.show() 
```
